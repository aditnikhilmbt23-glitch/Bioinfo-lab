```python
import os
from google.colab import files
from Bio import SeqIO, Entrez
Entrez.email='A.N.Other@example.com' #always tell NCBI (server for genbank db) who you are
mysearch = Entrez.esearch(db='nucleotide',term="[cytochrome oxidase]")
myrecord = Entrez.read(mysearch)
mysearch.close()
myids=myrecord["IdList"]
for id1 in myids:
  filename = id1 + ".gb"
  if not os.path.isfile(filename):
    net_handle = Entrez.efetch(
        db='nucleotide', id=id1, rettype="gb", retmode="text"
    )
    out_handle= open(filename, "w")
    out_handle.write(net_handle.read())
    out_handle.close()
    net_handle.close()
  gb_record = SeqIO.read(filename,"genbank")
  print(f"Features for record {id1}:")
  found_mRNA = False
  for feature in gb_record.features:
    print(f"  - {feature.type}")
    if feature.type == "mRNA":
      print(f"mRNA sequence for {id1}: {feature.extract(gb_record.seq)}")
      found_mRNA = True
      break
  if not found_mRNA:
    print(f"  - No mRNA feature found for record {id1}")
```
